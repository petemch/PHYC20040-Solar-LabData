```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
#B - V data from observations
BminusV = np.array([0.82, -0.11, 0.35, 0.55, 1.01, 1.01, 0.36, 0.99, 0.64, 0.02, 0.47, 1.18, -0.09, 1.23, 0.6, 1.15, 1.7, -0.07, 0.56, 0.12, 0.78, 1.15, 0.17, 0.13])
```

```python
#sample main sequence data
AbsMag= np.array([-5.8, -4.1, -1.1, -0.7, 2.0, 2.6, 3.4, 4.4, 5.1, 5.9, 7.3, 9.0, 11.8, 16])
SampleBminusV = np.array([-0.35, -0.31, -0.16, 0, 0.13, 0.27, 0.42, 0.58, 0.70, 0.89, 1.18, 1.45, 1.63, 1.80 ])
```

```python
#plotting/ aligning main sequences
plt.title('Plot of Aligned Main Sequences- best estimate for offset')
plt.xlabel('B-V')
plt.ylabel('Apparent/Absolute Magnitude (V)')
plt.plot(BminusV, V, 'r+', label='Data from Pleiades Cluster' )
plt.plot(SampleBminusV, AbsMag +5.5, label='Sample Main Sequence with added offset of 5.5')
plt.ylim(max(V), min(AbsMag))
plt.legend()
```
